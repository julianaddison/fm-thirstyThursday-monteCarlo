# Example I - Monty Hall Problem
---

In the Monty Hall problem, the TV show host Monty presents three doors to the participant. Behind one of the doors is a valuable prize like a car, while behind the others is a less valuable prize like a goat.  

Consider yourself to be one of the participants in the show. You choose one out of the three doors. Before opening your chosen door, Monty opens another door behind which would be one of the goats. Now you are left with two doors, behind one could be the car, and behind the other would be the other goat. 

Monty then gives you the option to either switch your answer to the other unopened door or stick to the original one.  

Is it in your favor to switch your answer to the other door? Well, probability says it is! Let's test this through Monte Carlo simulations!!
  

```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. Monte Carlo Function

```python
def monte_carlo(iterations, doors=['car', 'goat', 'goat']):

    # list to store probability values after each iteration
    switch_win_prob = []
    stick_win_prob = []

    # counting wins by switch or sticking
    win_switch_counter = 0
    win_stick_counter = 0

    for i in range(iterations):

        # Before starting the game, we randomize the prizes behind each door. 
        # One of the doors will have a car behind it, while the other two will have a goat each. 
        # When we play a large number of games, all possible permutations get covered of prize distributions,
        # and door choices get covered.
        
        # randomize the order of prizes behind each door
        random.shuffle(doors)

        # 2. select a door from either of the 3 doors
        k = random.randrange(0, 2)

        # The code below decides if your choice was correct or not,
        # and if switching wouldâ€™ve been the correct move.  
        if doors[k] != 'car':
            # if you don't get the car behind your choice switching 
		    # would have been the right move
            win_switch_counter += 1
        else:
            # if you won the car, then sticking would be the right choice
            win_stick_counter += 1

        # After playing each game, the winning probabilities are
        # updated and stored in the lists. When all games have been
        # played, we return the final values of each of the lists, 
        # i.e., winning by switching your choice and winning by 
        # sticking to your choice. 
        
        # 3. calculate probabilities
        win_by_switch = win_switch_counter/(i+1)
        win_by_stick = win_stick_counter/(i+1) 
 
        switch_win_prob.append(win_by_switch)
        stick_win_prob.append(win_by_stick)
    
    
    return switch_win_prob, stick_win_prob
```

# 2. Run Simulation

```python
# update number
NUM_ITERATIONS = 1000

# simulate iterations
switch_prob, stick_prob = monte_carlo(NUM_ITERATIONS)

# print probability values after all iterations
print(f'Always switching doors winning probability: {switch_prob[-1]}')
print(f'Always stick doors winning probability: {stick_prob[-1]}')
print()

# 4. plot and analyze probabilities as the number of iteration increase
plt.figure(figsize=(15,9))
plt.plot(switch_prob)
plt.plot(stick_prob)
plt.tight_layout()
plt.show()
```

After running the simulation 1000 times, the probability that we win by always switching is 67.7%, and the probability that we win by always sticking to our choice is 32.3%. In other words, you will win approximately 2/3 times if you switch your door, and only 1/3 times if you stick to the original door. Therefore, according to the Monte Carlo simulation, we are confident that it works to our advantage to switch the door in this tricky game.
